```python
##Check duplicates
print(df.duplicated().sum()) 
df.drop_duplicates(inplace=True)  # Remove duplicates if needed
```

```python
df["Market_Open"] = df["Market_Open"].astype(int)
df = df.apply(pd.to_numeric, errors='coerce')  # Convert all applicable columns to numeric
```

```python
# Weather data for US regions
us_regions = [
    "soux_H", "soux_L", "indianap_H", "indianap_L", "memphis_H", "memphis_L"
]

# Commodity price data
commodity_price = ["Gold", "USD", "Oil"]

# Soybean market data (Open, Close, High, Low, Volume)
soy_OCHLV = [
    "Soy_Bean_high", "Soy_Bean_low", "Soy_Bean_settle", "Soy_Bean_volume", "Soy_Bean_openint",
    "Soy_Meal_high", "Soy_Meal_low", "Soy_Meal_settle", "Soy_Meal_volume", "Soy_Meal_openint",
    "Soy_Oil_high", "Soy_Oil_low", "Soy_Oil_settle", "Soy_Oil_volume", "Soy_Oil_openint"
]

# Production data for different countries
production_origin = [
    "US_Area", "US_Production", "Brazil_Area", "Brazil_Production", "Argentina_Area", "Argentina_Production",
    "China_Area", "China_Production", "India_Area", "India_Production", "Paraguay_Area", "Paraguay_Production",
    "Canada_Area", "Canada_Production", "RussianF_Area", "RussianF_Production", "CentAmer_Area", "CentAmer_Production",
    "Bolivia_Area", "Bolivia_Production", "Africa_Area", "Africa_Production"
]
```

```python
plt.figure(figsize=(20, 6))
sns.kdeplot(df["soux_H"], label="SIOUX FALLS", shade=True)
sns.kdeplot(df["indianap_H"], label="INDIANAPOLIS", shade=True)
sns.kdeplot(df["memphis_H"], label="MEMPHIS", shade=True)
plt.title("HIGH TEMPERATURE DISTRIBUTION")
plt.xlabel("Temperature (°F)")
plt.ylabel("Density")
plt.legend()
plt.grid()

plt.figure(figsize=(20, 6))
sns.kdeplot(df["soux_L"], label="SIOUX FALLS", shade=True)
sns.kdeplot(df["indianap_L"], label="INDIANAPOLIS", shade=True)
sns.kdeplot(df["memphis_L"], label="MEMPHIS", shade=True)
plt.title("LOW TEMPERATURE DISTRIBUTION")
plt.xlabel("Temperature (°F)")
plt.ylabel("Density")
plt.legend()
plt.grid()

plt.show()
```

```python
# Calculate mid temperatures correctly
soux_mid_temp = ((df['soux_H'] - df['soux_L']) / 2) + df['soux_L']
indianap_mid_temp = ((df['indianap_H'] - df['indianap_L']) / 2) + df['indianap_L']
memphis_mid_temp = ((df['memphis_H'] - df['memphis_L']) / 2) + df['memphis_L']

# Calculate the average temperature
avg_temp = (soux_mid_temp + indianap_mid_temp + memphis_mid_temp) / 3

# Get unique years from the dataset
years = np.unique(df.index.year)

# Compute yearly temperature average and soybean production
yearly_temp_avg = [avg_temp.loc[str(i)].mean() for i in years]
US_soybean_prod = [df["US_Production"].loc[str(i)].mean() for i in years]

# Remove 1961 data (first year) due to missing temperature records
yearly_temp_avg.pop(0)
US_soybean_prod.pop(0)
years = years[1:]  # Adjust the years array

# Scatter Plot: Temperature vs. Soybean Production
plt.figure(figsize=(14, 14))
plt.title("TEMPERATURE AND SOYBEAN PRODUCTION")
plt.scatter(yearly_temp_avg, US_soybean_prod)
plt.plot(yearly_temp_avg, US_soybean_prod, alpha=0.3)
plt.xlabel("Average Temperature (°F)")
plt.ylabel("Soybean Production (Bushels)")
plt.grid()
plt.show()

# Dual-Axis Plot: Temperature and Production Over Years
fig, ax1 = plt.subplots(figsize=(15, 6))
ax1.set_title("TEMPERATURE AND SOYBEAN PRODUCTION OVER YEARS")
ax1.plot(years, yearly_temp_avg, '-r')
ax1.set_xlabel('Year')
ax1.set_ylabel('Temperature (°F)')
ax1.tick_params('y')
ax1.legend(["Temperature"], loc=1)

ax2 = ax1.twinx()
ax2.plot(years, US_soybean_prod)
ax2.set_ylabel('Soybean Production (Bushels)')
ax2.tick_params('y')
ax2.legend(["Production"], loc=0)

ax1.grid()
plt.show()
```

```python
plt.figure(figsize=(20, 18))

average_production = []

# Scatter plot for each country's area vs production
for i in range(0, len(production_origin), 2):  
    plt.scatter(
        np.log10(df[production_origin[i]]),  # Area (log scale)
        np.log10(df[production_origin[i + 1]])  # Production (log scale)
    )
    average_production.append(df[production_origin[i + 1]].mean())  # Append mean production

# Compute the overall average production
avg_production = int(np.mean(average_production))

# Title and Labels
plt.grid()
plt.title(f"Number of Area and Production (Avg: {avg_production})")
plt.xlabel("Area of Soybean Plantation - Log Scale")
plt.ylabel("Production of Soybean Plantation - Log Scale")

# Legend for each country
plt.legend([
    "US", "Brazil", "Argentina", "China",
    "India", "Paraguay", "Canada",
    "Russia", "Central America", "Bolivia", "Africa"
])

plt.show()
```

```python
# First loop: Area vs Production for each country
for i in range(0, len(production_origin), 2):  
    fig, ax1 = plt.subplots(figsize=(15, 6))
    
    country = production_origin[i].split("_")[0]  # Extract country name
    
    ax1.set_title(f"{country} AREA AND SOYBEAN PRODUCTION")
    ax1.plot(df[production_origin[i]], '-r')  # Area
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Area')
    ax1.tick_params('y')
    ax1.legend(["AREA"], loc=0)

    # Twin axis for Production
    ax2 = ax1.twinx()
    ax2.plot(df[production_origin[i + 1]], '-b')  # Production
    ax2.set_ylabel('Soybean Production')
    ax2.tick_params('y')
    ax2.legend(["PRODUCTION"], loc=4)

    ax1.grid()
    plt.show()  # Show each plot separately

# Second plot: Land Efficiency (Production per Unit Area)
plt.figure(figsize=(15, 8))

for i in range(0, len(production_origin), 2):  
    efficiency = df[production_origin[i + 1]] / df[production_origin[i]].replace(0, np.nan)  
    plt.plot(efficiency)

plt.ylabel("Efficiency (Production / Area)")
plt.xlabel("Year")
plt.title("LAND EFFICIENCY")

# Legend for each country
plt.legend([
    "US", "Brazil", "Argentina", "China",
    "India", "Paraguay", "Canada",
    "Russia", "Central America", "Bolivia", "Africa"
])

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot commodity prices over time
plt.figure(figsize=(14, 6))
plt.title("Commodities Price Over Time")
plt.plot(df['Gold'], label="Gold")
plt.plot(df['USD'], label="USD")
plt.plot(df['Oil'], label="Oil")
plt.plot(df['bean_settle'], label="Soybean")
plt.xlabel("Year")
plt.ylabel("Price")
plt.grid()
plt.legend()
plt.show()

# Function to plot percentage change over time
def plot_long(series, title):
    plt.figure(figsize=(14, 3))
    plt.title(title)
    plt.xlabel("Year")
    plt.ylabel("Price Change (%)")
    plt.grid()
    plt.plot(series.dropna())  # Drop NaN from percentage change
    plt.show()

# Plot percentage change for each commodity
plot_long(df['Gold'].pct_change(), "Gold Price Change in Percent Over Time")
plot_long(df['USD'].pct_change(), "USD Price Change in Percent Over Time")
plot_long(df['Oil'].pct_change(), "Oil Price Change in Percent Over Time")
plot_long(df['bean_settle'].pct_change(), "Soybean Price Change in Percent Over Time")
```

```python
years = np.unique(df.index.year)
average_soybean_price = [df["bean_settle"].loc[str(i)].mean() for i in years]
total_soybean_production = [df[[production_origin[i+1] for i in range(0,22,2)]].sum(axis=1).loc[str(i)].mean() for i in years]

fig, ax1 = plt.subplots(figsize=(15,6))
ax1.set_title("PRICE AND SOYBEAN PRODUCTION")
ax1.plot(years, average_soybean_price, '-r')
ax1.set_xlabel('Year')
ax1.set_ylabel('Price')
ax1.tick_params('y')
ax1.legend(["PRICE"], loc=2)

ax2 = ax1.twinx()
ax2.plot(years, total_soybean_production)
ax2.set_ylabel('Soybean Production')
ax2.tick_params('y')
ax2.legend(["PRODUCTION"], loc=4)
ax1.grid()
ax2.grid(linestyle=":")

plt.figure(figsize=(15,6))
plt.title("Soybean Demand")
plt.xlabel("Year")
plt.ylabel("Demand")
plt.grid()
plt.plot(years, np.multiply(total_soybean_production, average_soybean_price))
```
